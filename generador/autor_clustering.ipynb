### Generación del Clustering de Autores

```python
import pandas as pd
from sklearn.preprocessing import MinMaxScaler

from sklearn.cluster import KMeans
```

Cargamos y quitamos las columnas que no usaremos

```python
df = pd.read_csv('autors.csv')
df.drop(columns=['Unnamed: 0', 'authors', 'Average Minutes to Read', 'Most Common Writting Style'], inplace =True)
```

```python
df.columns
```

La función preprocess_dataframe reescala las columnas numéricas entre 0 y 1 y realiza dummies sobre las categóricas

```python
def preprocess_dataframe(df):
    scaler = MinMaxScaler()

    numeric_cols = df.select_dtypes(include=['number']).columns
    categorical_cols = df.select_dtypes(include=['object', 'category']).columns

    df[numeric_cols] = scaler.fit_transform(df[numeric_cols])
    print(df.shape)

    df = pd.get_dummies(df, columns=categorical_cols, drop_first=True)

    return df
```

```python
df = preprocess_dataframe(df)
```

```python
df.columns
```

Generamos la lista con los coeficientes para cada una de las variables

```python
a = [0.5, 1.5, 1.2, 1.2, 1.5] + [2]*5 + [2.5]*17

```

```python
df
```

```python
df = df.apply(pd.to_numeric, errors='coerce')
```

```python
df.multiply(a, axis = 1)
```

Realizamos el cluster de k_means con 10 clusters

```python
num_clusters = 10

kmeans = KMeans(n_clusters=num_clusters, random_state=0).fit(df)

labels = kmeans.labels_
centers = kmeans.cluster_centers_

df['Cluster'] = labels
```

```python
df['Cluster'].value_counts()
```

```python
df_original = pd.read_csv('autors.csv')
```

Juntamos los resultados con el dataset original

```python
df_con = pd.concat([df_original, df['Cluster']], axis = 1)
```

```python
df_con
```

```python
df_con.to_csv('authors_on_cluster.csv')
```
